```python
import numpy as np
import numpy.random
import matplotlib
import matplotlib.pyplot as plt
```

```python
def city(coordinate, population, variance):
    var = random.normal(variance)
    cov = [[var,0],[0,var]]
    return random.multivariate_normal(coordinate,cov,population).T
```

```python
def random_sampling(n):
    return n * random.uniform()
```

```python
def cities(cities, world_dimensions):
    return world_dimensions*random.rand(cities,2)
```

```python
def plot_cities(n_cities, max_size, max_variance, world_dimensions):
    plt.figure(figsize = (13,13))
    world_cities = cities(n_cities, world_dimensions)
    for c in world_cities:
        population = int(random_sampling(max_size))
        variance = random_sampling(max_variance)
        cov = [[variance,0],[0,variance]]
        x,y = random.multivariate_normal(c,cov,population).T
        plt.scatter(x,y,s=3)
```

```python
plot_cities(30,1000,2000,1600)
```

```python
class Human():
    def __init__(self, dimensions):
        self.dimensions = dimensions
        self.x = random.randint(0, dimensions+1)
        self.y = random.randint(0, dimensions+1)
        self.speed_x = random.randint(-1,2)
        self.speed_y = random.randint(-1,2)
                
    def accelerate(self):
        self.speed_x += random.randint(-1,2)
        self.speed_y += random.randint(-1,2)
        
    def move(self):
        self.accelerate()
        self.x += self.speed_x
        self.y += self.speed_y
        
    def check_edge(self):
        if self.x == 0:
            self.x += 1
        if self.x == self.dimensions:
            self.x -= 1
        if self.y == 0:
            self.y += 1
        if self.y == self.dimensions:
            self.y -= 1
```

```python
population = [Human(30) for i in range(10)]
```

```python
plt.figure(figsize = (7,6))
for p in population:
    p.move()
    plt.scatter(p.x,p.y)
    p.check_edge
```
